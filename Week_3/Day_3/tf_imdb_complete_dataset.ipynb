```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data/movie_metadata.csv')
df
```

```python
input_data = df[['num_critic_for_reviews','duration','gross','budget','imdb_score']]
```

```python
input_data.head()
```

```python
# print(max([max(budget) for budget in input_data]))
# print(input_data['budget'])
input_data['budget'] = input_data['budget'].fillna(0)
input_data['gross'] = input_data['gross'].fillna(0)
input_data['budget'] = input_data['budget'].fillna(0)
input_data['duration'] = input_data['duration'].fillna(0)
input_data['imdb_score'] = input_data['imdb_score'].fillna(0)
input_data['num_critic_for_reviews'] = input_data['num_critic_for_reviews'].fillna(0)
```

```python
input_data.head()
input_data.sample(frac=1)
input_data['gross'] /= input_data['gross'].max()
input_data['duration'] /= input_data['duration'].max()
input_data['num_critic_for_reviews'] /= input_data['num_critic_for_reviews'].max()
input_data['budget'] /= input_data['budget'].max()
input_data
```

```python
train_data = input_data[:4000]
train_label = train_data['imdb_score']
train_label = train_label/10
# train_data.drop('imdb_score',axis=1)
train_data = train_data[['num_critic_for_reviews','duration','gross','budget']]
# train_data.apply(lambda x: x/x.max(), axis=0)

test_data = input_data[4000:]
test_label = test_data['imdb_score']
test_label = test_label/10

# test_data.drop('imdb_score',axis=1)
test_data = test_data[['num_critic_for_reviews','duration','gross','budget']]

# train_data
# train_label.shape
```

```python
train_data_to_pass = train_data.values
test_data_to_pass = test_data.values
# type(train_data_to_pass)
```

```python
#Creating the model
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(512, activation=tf.nn.sigmoid, input_shape=(4,)))
model.add(tf.keras.layers.Dense(1, activation=tf.nn.sigmoid))
```

```python
model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.summary()
```

```python
model.fit(train_data_to_pass, train_label, epochs=5)
```

```python
#Test the model
loss, accuracy = model.evaluate(test_data_to_pass, test_label)
print('Test accuracy: %.4f' % (accuracy))
```
